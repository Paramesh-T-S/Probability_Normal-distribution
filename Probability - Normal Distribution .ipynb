```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
mu = 0
sigma = 1
```

```python
x = np.linspace(-3.5, 3.5, 5000)

y_pdf = ss.norm.pdf(x, mu, sigma)

plt.plot(x, y_pdf, label = 'pdf')
```

```python
mu = 5
sigma = 3
```

```python
mean, variance, skewness, kurtosis = ss.norm.stats(mu, sigma, moments='mvsk')
```

```python
print("Mean = {}".format(mean))
print("Median = {}".format(ss.norm.median(mu,sigma)))
print("Varianve = {}".format(variance))
print("Skewness = {}".format(skewness))
print("Kurtosis = {}".format(kurtosis))
```

```python
alpha = 0.1
percent = ss.norm.ppf(alpha, mu, sigma)
print("Percentile = {}".format(percent))
```
